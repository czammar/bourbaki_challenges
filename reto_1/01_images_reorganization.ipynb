```python
import os
import pandas as pd

from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader, Dataset
from sklearn.model_selection import train_test_split

from PIL import Image
import shutil
```

```python
#pip install pytest-shutil
```

```python
y_train_ids = pd.read_csv(
    "./data/Datos-del-Reto-I/y_train_ids.csv",
    sep=','
    )
```

```python
y_train_ids.sample(10)
```

```python
labeling_map = dict(zip(y_train_ids.images, y_train_ids.labels))
```

```python
labeling_map
```

```python

X_train_, X_test, y_train_, y_test = train_test_split(
    y_train_ids["images"],
    y_train_ids["labels"], 
    test_size=0.20,
    random_state=123,
    stratify=y_train_ids["labels"],
    shuffle=True
    )
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    X_train_,
    y_train_, 
    test_size=0.20,
    random_state=123,
    stratify=y_train_,
    shuffle=True
    )
```

```python
if not os.path.exists(os.path.join(os.getcwd(),'data\\custom_data\\train')): 
      
    # if the demo_folder directory is not present  
    # then create it. 
    os.makedirs(os.path.join(os.getcwd(),'data\\custom_data\\train')) 
```

```python
if not os.path.exists(os.path.join(os.getcwd(),'data\\custom_data\\val')): 
      
    # if the demo_folder directory is not present  
    # then create it. 
    os.makedirs(os.path.join(os.getcwd(),'data\\custom_data\\val')) 
```

```python
if not os.path.exists(os.path.join(os.getcwd(),'data\\custom_data\\test')): 
      
    # if the demo_folder directory is not present  
    # then create it. 
    os.makedirs(os.path.join(os.getcwd(),'data\\custom_data\\test')) 
```

```python
images_reorganized = pd.concat(
    [
        pd.DataFrame(X_train).assign(group="train"),
        pd.DataFrame(X_val).assign(group="val"),
        pd.DataFrame(X_test).assign(group="test")
    ]
).reset_index(drop=True)

```

```python
images_reorganized['label'] = images_reorganized.apply(lambda x: labeling_map[x['images']], axis=1)
```

```python
images_reorganized['folder_label'] = images_reorganized['label'] + 1 
```

```python
images_reorganized.head(5)
```

```python
MAIN_PATH = os.path.join(os.getcwd(),'data\\Data_challenge\\x_train')
```

```python
for index, row in images_reorganized.iterrows():

    class_label = str(round(row['label'], 0))
    folder_label = str(round(row['folder_label'],0))
    group = row['group']
    image = row['images']

    jpgfile = os.path.join(os.getcwd(),'data\\Data_challenge\\x_train\\', folder_label, image)
    dst_dir = os.path.join(os.getcwd(),'data\\custom_data\\', group, image)    

    shutil.copy(jpgfile, dst_dir)
```

```python
images_reorganized.to_csv(
    os.path.join(os.getcwd(),'data\\custom_data\\', "images_reorganization.csv")    ,
    index=False
    )
```
