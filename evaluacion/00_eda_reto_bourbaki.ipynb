```python
images_list = []

for _, row in y_train[["id_annonce"]].iterrows():

  ID_ANNOUNCE = row["id_annonce"]
  PATH_IMAGE = f'./reduced_images/train/ann_{ID_ANNOUNCE}/'

  #print(row["id_annonce"])

  file_list=[]

  for file in os.listdir(PATH_IMAGE):
    if file.endswith('.jpg'):
      file_list.append(file)

  temp_df = pd.DataFrame({"image_name": file_list})
  temp_df["path_image"] = PATH_IMAGE
  temp_df["id_annonce"] = ID_ANNOUNCE

  images_list.append(temp_df)
```

```python
data_images = pd.concat(images_list)
```

```python
data_images["long_path_image"] = data_images["path_image"]+ data_images["image_name"]
```

```python
data_images['total_images']=data_images.groupby('id_annonce')['path_image'].transform('count')
```

```python
data_images = data_images[["id_annonce", "path_image", "image_name", "long_path_image", "total_images"]]
```
